```python
"""
SELECT age, gender, education
  FROM players
  INTO OUTFILE 'C:/ProgramData/MySQL/MySQL Server 8.0/Uploads/demo.txt';
"""
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
ages = []
noage = 0
gends = []
nogend = 0
edus = []
noedu = 0
with open('demo.txt') as f:
    for line in f:
        age,gend,edu = line.split('\t')
        age = int(age)
        gend = int(gend)
        edu = int(edu)
        
        if age>=0:
            ages.append(age)
        else:
            noage += 1
        if gend>=0:
            gends.append(gend)
        else:
            nogend += 1
        if edu>=0:
            edus.append(edu)
        else:
            noedu += 1
        
```

```python
fig,ax = plt.subplots()
labels = ['prefer not', '<7', '7-12', '12-17', '18-25', '26-35', '36-45', '46-55', '56-65', '66-75', '>76']
sns.distplot(ages, ax=ax, bins=len(labels)-1)
ax.set_xticks(np.arange(0, len(labels))+.5)
ax.set_xticklabels(labels, rotation=90)
print(noage)
```

```python
fig,ax = plt.subplots()
labels = ['prefer not', 'female', 'male', 'other']
sns.distplot(gends, ax=ax, bins=len(labels)-1, kde=False)
ax.set_xticks(np.arange(0, len(labels))+.5)
ax.set_xticklabels(labels, rotation=90)
print(nogend)
```

```python
fig,ax = plt.subplots()
labels = ['prefer not', 'primary', 'secondary', 'A-level', 'Bachelors', 'Masters+']
sns.distplot(edus, ax=ax, bins=len(labels)-1, kde=False)
ax.set_xticks(np.arange(0, len(labels))+.5)
ax.set_xticklabels(labels, rotation=90)
print(noedu)
```
